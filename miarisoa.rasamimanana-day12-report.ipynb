# Exercise 1:

Create a simple line plot for y = x for x values ranging from 0 to 10.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
#inputs
x = np.arange(0,11)
y=x

# plot values
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title(r"Plot of $y=x$")
plt.show()
```

# Exercise 8:

Add titles, axis labels, and a legend to the plot in Exercise 7.

```python
#inputs
x = np.arange(0,11)
#plots
fig, ax = plt.subplots(figsize=(8,4), dpi=100)
ax.plot(x, x**2,color='blue', label=r"y = x^2")
ax.plot(x, x**3,color='red', label=r"y = x^3")
#labels and legend
ax.legend(loc=2) 
ax.set_xlabel(r'x', fontsize=18)
ax.set_ylabel(r'y', fontsize=18)
ax.set_title(r'Plot of y=x^2 and y=x^3')
plt.show()
```

# Exercise 9:

Create a subplot with 2 rows and 1 column. Plot $y1 = x^2$ in the first subplot and $y2 = x^3$ in the second.

# Exercise 19:

Add error bars to a bar plot with categories = ['A', 'B', 'C', 'D'] and values = [3, 7, 5, 9], assuming the error is 1 for each value.

# Exercise 20:

Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.

```python
# inputs 
x = np.linspace(0, 2 * np.pi, endpoint=True)
y_S = np.sin(x)
y_C = np.cos(x)

# create the plot
plt.figure(figsize=(8, 4))

# sine wave
plt.plot(x, y_S, linestyle='--', color='b', label='Sine Wave')

# cosine wave 
plt.plot(x, y_C, linestyle='-', color='r', label='Cosine Wave')


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

# Exercise 21:

Plot a contour plot for the function $z = \sin(x) + \cos(y)$ over a grid of x and y values ranging from -5 to 5.

```python
#function
def f(x,y):
    return np.sin(x) + np.cos(y)
#inputs
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# create the contour plot
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, f(X,Y), 8, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X,Y), 8, colors='black')

plt.clabel(C, inline=1, fontsize=10)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

# Exercise 26:

Create a 3D plot of the function $z = x^2 + y^2$ over a grid of x and y values.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# create figure and set figure size and dpi (dots per inch)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# inputs
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)

# 3D surface
Z = X**2 + Y**2

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)

# Show the plot
plt.show()
```

# Exercise 34 :

Create a plot with twin y-axes where one axis represents y = x and the other represents $y = x^2$.

```python
#inputs
x = np.linspace(0, 10, 100)
y1 = x          
y2 = x**2

fig, ax1 = plt.subplots(figsize=(10, 6))

#first axis, plot y = x
ax1.plot(x, y1, color="red", label='y = x')  
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y = x', color='b')
ax1.tick_params(axis='y', labelcolor='b')

#second y-axis, plot y = x^2
ax2 = ax1.twinx()  
ax2.plot(x, y2,  color="blue", label='y = x^2') 
ax2.set_ylabel('Y = x^2', color='r')
ax2.tick_params(axis='y', labelcolor='r')


plt.grid()
plt.show()
```
